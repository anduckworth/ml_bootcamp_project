```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('vader_lexicon')
import os
import numpy as np
import re
from dotenv import load_dotenv
load_dotenv()
sentiment_analyzer = SentimentIntensityAnalyzer()
```

```python
def tweet_sentiment(user):
    
    maxTweets=10000
    tweets_list = []
     
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:'+ user).get_items()):
        if i>maxTweets:
            break
        tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
   
    df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
    
    df['Datetime'] = pd.to_datetime(df['Datetime'], format='%Y-%m-%d')
    df['Datetime'] = df['Datetime']
    df.set_index('Datetime',inplace=True)
    df = df.loc[:'2021-01-01']
    
    
          
    sentiments = []

    for i in range(len(df)):
        try:
            text = df["Text"][i]
            sentiment = sentiment_analyzer.polarity_scores(text)
            compound = sentiment["compound"]
            pos = sentiment["pos"]
            neu = sentiment["neu"]
            neg = sentiment["neg"]

            sentiments.append({
                "compound": compound,
                "positive": pos,
                "negative": neg,
                "neutral": neu

            })

        except AttributeError:
            pass
    
    
    df_sent = pd.DataFrame(sentiments)
    df = df.reset_index()
    df_sentiment = pd.concat([df, df_sent], axis=1)
    df_sentiment = df_sentiment.set_index('Datetime')
    return df_sentiment
```

```python
mike = tweet_sentiment('michael_saylor')
elon = tweet_sentiment('elonmusk')
bitboy = tweet_sentiment('Bitboy_Crypto')
woo = tweet_sentiment('woonomic')
t360 = tweet_sentiment('360_trader')
cnbc = tweet_sentiment('CNBC')
altcoin = tweet_sentiment('AltcoinDailyio')
coindesk = tweet_sentiment('CoinDesk')
pent = tweet_sentiment('Pentosh1')
whale = tweet_sentiment('CryptoWhale')
charles = tweet_sentiment('IOHK_charles')
mark = tweet_sentiment('mcuban')
kevin = tweet_sentiment('kevinolearytv')
cage = tweet_sentiment('Kevin_Cage_')
joel = tweet_sentiment('JoelKatz')
dave = tweet_sentiment('stoolpresidente')
tyler = tweet_sentiment('tyler')
cameron = tweet_sentiment('cameron')
```
